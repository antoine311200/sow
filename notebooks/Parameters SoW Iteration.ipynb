```python
import torch
import numpy as np

import matplotlib.pyplot as plt
```

```python
n, m = 512, 1376

# rcParams dpi
plt.rcParams['figure.dpi'] = 200

r = np.linspace(1, 50, 50)
N = [2, 3, 5, 10, 15, 20]

colors = plt.cm.viridis(np.linspace(0, 0.8, len(N)))

plt.fill_between(r, n*m, 2*(n+m)*r, color='green', alpha=0.065)
for i, p in enumerate(N):
    plt.plot(r, p*(n+m)*r, label=f'n_iter={p}', color=colors[i], lw=0.8)
    
    for rk in r[::10]:
        rank = min(n, m, int(p*rk))
        plt.scatter(rk, p*(n+m)*rk, color=colors[i], s=2, marker='o')
        plt.text(rk+1, p*(n+m)*rk*0.95, '$r_{'+str(int(rank))+'}$', color=colors[i], fontdict={'size': 8})
plt.axhline(n*m, label='Original weight', color='red', linestyle='--', lw=0.8)
plt.text(r[int(len(r)*0.1)], n*m*1.1, '$r_{'+str(min(n,m))+'}$', color='red', fontdict={'size': 8})
plt.legend(fontsize=8)
plt.yscale('log')
plt.xlabel('Ranks')
plt.xlim(1, 50)
# plt.ylim(1e4, 1e7)
plt.ylabel('Number of parameters')
plt.title('Number of parameters of sum-rank decomposition')
plt.savefig('sum-rank-decomposition.png', dpi=200, transparent=True)
plt.show()
```
